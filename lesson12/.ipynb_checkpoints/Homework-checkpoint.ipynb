```python
import os
import cv2
from time import time
import numpy as np
from sklearn.utils import shuffle

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 6]
```

```python
folder = 'data/subset_homework'

# Load traffic sign class 0
fnames_0 = os.listdir(os.path.join(folder, 'class_id_0'))
images_0 = [cv2.imread(os.path.join(folder, 'class_id_0', f), cv2.IMREAD_UNCHANGED) for f in fnames_0]
labels_0 = [0] * len(images_0)

for cnt, idx in enumerate(np.random.randint(0, len(images_0), 5)):
    plt.subplot(1,5,cnt+1)
    plt.imshow(images_0[idx], cmap='gray', vmin=0, vmax=255)
    plt.title(labels_0[idx]), plt.axis(False)
```

```python
# Load traffic sign class 1
fnames_1 = os.listdir(os.path.join(folder, 'class_id_1'))
images_1 = [cv2.imread(os.path.join(folder, 'class_id_1', f), cv2.IMREAD_UNCHANGED) for f in fnames_1]
labels_1 = [1] * len(images_1)

for cnt, idx in enumerate(np.random.randint(0, len(images_1), 5)):
    plt.subplot(1,5,cnt+1)
    plt.imshow(images_1[idx], cmap='gray', vmin=0, vmax=255)
    plt.title(labels_1[idx]), plt.axis(False)
```

```python
print('Num samples class_0', len(images_0))
print('Num samples class_1', len(images_1))
```

```python
images = images_0 + images_1
labels = labels_0 + labels_1
images, labels = shuffle(images, labels)

images = np.array(images)
labels = np.array(labels)

print('Images', images.shape)
print('Labels', labels.shape)
```

```python
pixels = np.array([image.flatten() for image in images])/255

print('Shape', pixels.shape)
```

```python
for cnt, idx in enumerate(np.random.randint(0, len(images), 5)):    
    plt.subplot(1,5,cnt+1)
    plt.imshow(images[idx], cmap='gray', vmin=0, vmax=255)
    plt.title(labels[idx])
```

```python
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense, Input
```

The following netowork will consist of only one single neuron. It is a very tiny network (not even a network, strictly speaking :-) ) and yet it can be quite powerful.

```python
inputs = Input(shape=(pixels.shape[1]))
outputs = Dense(1, activation="linear")(inputs)
model = Model(inputs, outputs)

model.compile(optimizer ='adam', loss = 'mean_squared_error')

model.summary()
```

```python
tf.keras.utils.plot_model(model, to_file="model.png", show_shapes=True, show_layer_names=True)
```

```python
history = model.fit(pixels, labels, epochs=10, batch_size=32)
```

#### Performance Evaluation

Once our model is trained, we will can run it on our images to see how it performs (inference).

```python
# Run it on the entire dataset
predictions = model.predict(pixels).squeeze()
predictions = predictions > 0.5
```

```python
correct = 0
for prediction, label in zip(predictions, labels):
    if prediction == label:
        correct = correct + 1

print('Accuracy', correct/len(labels))
```

```python
# lets add several dense layers


inputs2 = Input(shape=(pixels.shape[1]))
outputs2 = Dense(5, activation="linear")(inputs2)
outputs3 = Dense(1, activation="relu")(outputs2)
model2 = Model(inputs2, outputs3)

model2.compile(optimizer ='adam', loss = 'mean_squared_error')

model2.summary()

history2 = model2.fit(pixels, labels, epochs=20, batch_size=32)
```

```python
tf.keras.utils.plot_model(model2, to_file="model.png", show_shapes=True, show_layer_names=True)
```

```python
# Run it on the entire dataset
predictions2 = model2.predict(pixels).squeeze()
predictions2 = predictions2 > 0.5

correct2 = 0
for prediction, label in zip(predictions2, labels):
    if prediction == label:
        correct2 = correct2 + 1

print('Accuracy', correct2/len(labels))
```
